```python
import torch
import torch.nn as nn
import torch.optim as optim
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

# Device configuration
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

print("using device: " + str(device))

# Hyperparameters
vocab_size = None
embed_dim = 256
num_heads = 8
hidden_dim = 512
num_layers = 4
dropout = 0.1
sequence_length = 100
k = 50
batch_size = 32
num_epochs = 1000
learning_rate = 0.001

# Function to load text from file
def load_text_from_file(filename):
    with open(filename, 'r', encoding='utf-8') as f:
        text = f.read().strip()
    return text

# Build vocabulary and token mappings
def build_vocab(text):
    tokens = text.split()
    token_counts = Counter(tokens)
    vocab = {token: idx for idx, (token, _) in enumerate(token_counts.most_common())}
    vocab['<pad>'] = len(vocab)
    vocab['<unk>'] = len(vocab)
    idx_to_token = {idx: token for token, idx in vocab.items()}
    return vocab, idx_to_token

# Convert text to token IDs
def text_to_ids(text, vocab):
    tokens = text.split()
    return [vocab.get(token, vocab['<unk>']) for token in tokens]

# Create sequences for training
def create_sequences(token_ids, sequence_length):
    total_length = sequence_length + 1
    sequences = []
    for i in range(len(token_ids) - total_length + 1):
        seq = token_ids[i:i + total_length]
        sequences.append(seq)
    return np.array(sequences)

# Load the text data
text_data = load_text_from_file('training_text.txt')  # Ensure this file exists

# Build vocabulary and prepare data
vocab, idx_to_token = build_vocab(text_data)
vocab_size = len(vocab)
token_ids = text_to_ids(text_data, vocab)
sequences = create_sequences(token_ids, sequence_length)
sequences = torch.tensor(sequences, dtype=torch.long)

# Split into input and target
train_data = sequences[:, :-1]
train_targets = sequences[:, 1:]

# Create DataLoader
train_dataset = torch.utils.data.TensorDataset(train_data, train_targets)
train_loader = torch.utils.data.DataLoader(train_dataset, batch_size=batch_size, shuffle=True)

# Positional Encoding
class PositionalEncoding(nn.Module):
    def __init__(self, sequence_length, embed_dim):
        super().__init__()
        pe = torch.zeros(sequence_length, embed_dim)
        position = torch.arange(0, sequence_length, dtype=torch.float).unsqueeze(1)
        div_term = torch.exp(torch.arange(0, embed_dim, 2).float() * (-np.log(10000.0) / embed_dim))
        pe[:, 0::2] = torch.sin(position * div_term)
        pe[:, 1::2] = torch.cos(position * div_term)
        pe = pe.unsqueeze(0)
        self.register_buffer('pe', pe)

    def forward(self, x):
        return x + self.pe[:, :x.size(1), :]

# Custom Self-Attention Layer
class SelfAttention(nn.Module):
    def __init__(self, embed_dim, num_heads):
        super().__init__()
        assert embed_dim % num_heads == 0, "embed_dim must be divisible by num_heads"
        self.embed_dim = embed_dim
        self.num_heads = num_heads
        self.head_dim = embed_dim // num_heads
        
        self.query_proj = nn.Linear(embed_dim, embed_dim)
        self.key_proj = nn.Linear(embed_dim, embed_dim)
        self.value_proj = nn.Linear(embed_dim, embed_dim)
        self.out_proj = nn.Linear(embed_dim, embed_dim)
        self.dropout = nn.Dropout(dropout)
        self.scale = torch.sqrt(torch.tensor(self.head_dim, dtype=torch.float32))

    def forward(self, x, mask=None, return_attn=False):
        batch_size, seq_len, _ = x.size()
        
        Q = self.query_proj(x)
        K = self.key_proj(x)
        V = self.value_proj(x)
        
        Q = Q.view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)
        K = K.view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)
        V = V.view(batch_size, seq_len, self.num_heads, self.head_dim).transpose(1, 2)
        
        scores = torch.matmul(Q, K.transpose(-2, -1)) / self.scale
        
        if mask is not None:
            scores = scores.masked_fill(mask == 0, float('-inf'))
        
        attn_weights = torch.softmax(scores, dim=-1)
        attn_weights = self.dropout(attn_weights)
        
        attn_output = torch.matmul(attn_weights, V)
        attn_output = attn_output.transpose(1, 2).contiguous().view(batch_size, seq_len, self.embed_dim)
        output = self.out_proj(attn_output)
        
        if return_attn:
            return output, attn_weights
        return output

# GPT-like Transformer Model
class GPTLikeModel(nn.Module):
    def __init__(self, vocab_size, embed_dim, num_heads, hidden_dim, num_layers, dropout, sequence_length):
        super().__init__()
        
        self.token_embedding = nn.Embedding(vocab_size, embed_dim)
        self.positional_encoding = PositionalEncoding(sequence_length, embed_dim)
        self.num_heads = num_heads
        
        self.layers = nn.ModuleList([
            nn.ModuleDict({
                'self_attention': SelfAttention(embed_dim, num_heads),
                'norm1': nn.LayerNorm(embed_dim, eps=1e-6),
                'ffn': nn.Sequential(
                    nn.Linear(embed_dim, hidden_dim),
                    nn.ReLU(),
                    nn.Dropout(dropout),
                    nn.Linear(hidden_dim, embed_dim)
                ),
                'norm2': nn.LayerNorm(embed_dim, eps=1e-6),
                'dropout': nn.Dropout(dropout)
            }) for _ in range(num_layers)
        ])
        
        self.output_layer = nn.Linear(embed_dim, vocab_size)
        self.sequence_length = sequence_length

    def forward(self, x, return_attn=False):
        batch_size = x.size(0)
        x = self.token_embedding(x)
        x = self.positional_encoding(x)
        
        mask = torch.tril(torch.ones(x.size(1), x.size(1))).unsqueeze(0).unsqueeze(0).to(x.device)
        mask = mask.expand(batch_size, self.num_heads, -1, -1)
        
        attn_outputs = []
        for layer in self.layers:
            attn_output, attn_weights = layer['self_attention'](x, mask=mask, return_attn=True)
            attn_output = layer['dropout'](attn_output)
            x = layer['norm1'](x + attn_output)
            ffn_output = layer['ffn'](x)
            ffn_output = layer['dropout'](ffn_output)
            x = layer['norm2'](x + ffn_output)
            if return_attn:
                attn_outputs.append(attn_weights)
        
        output = self.output_layer(x)
        if return_attn:
            return output, attn_outputs
        return output

# Training Function
def train_model(model, train_loader, criterion, optimizer, num_epochs):
    model.train()
    losses = []
    
    for epoch in range(num_epochs):
        epoch_loss = 0
        for batch_idx, (inputs, targets) in enumerate(train_loader):
            inputs, targets = inputs.to(device), targets.to(device)
            
            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs.view(-1, vocab_size), targets.view(-1))
            loss.backward()
            optimizer.step()
            
            epoch_loss += loss.item()
        
        avg_loss = epoch_loss / len(train_loader)
        losses.append(avg_loss)
        print(f"\rEpoch [{epoch+1}/{num_epochs}], Loss: {avg_loss:.4f}", end=" ")
    
    return losses

# Plot Loss
def plot_loss(losses):
    plt.plot(range(1, len(losses) + 1), losses, label="Training Loss")
    plt.xlabel("Epoch")
    plt.ylabel("Loss")
    plt.title("Training Loss Over Time")
    plt.legend()
    plt.show()

# Generate k words and return sequence
def generate_text(model, start_sequence, vocab, idx_to_token, k, sequence_length):
    model.eval()
    tokens = start_sequence.split()
    input_ids = [vocab.get(word, vocab['<unk>']) for word in tokens]
    padding_needed = sequence_length - len(input_ids)
    if padding_needed > 0:
        input_ids = [vocab['<pad>']] * padding_needed + input_ids
    elif padding_needed < 0:
        input_ids = input_ids[-sequence_length:]
    input_ids = torch.tensor(input_ids, dtype=torch.long).unsqueeze(0).to(device)
    
    generated = []
    with torch.no_grad():
        for _ in range(k):
            outputs = model(input_ids)
            next_token_logits = outputs[:, -1, :]
            next_token = torch.argmax(next_token_logits, dim=-1)
            generated.append(next_token.item())
            input_ids = torch.cat([input_ids, next_token.unsqueeze(0)], dim=1)
            if input_ids.size(1) > sequence_length:
                input_ids = input_ids[:, 1:]
    
    generated_text = ' '.join([idx_to_token[token] for token in generated])
    return generated_text, input_ids

# Plot Vocabulary Influence (via Wo)
def plot_vocab_influence(model, vocab, idx_to_token, num_tokens=20):
    model.eval()
    with torch.no_grad():
        Wo = model.output_layer.weight.data.cpu().numpy()
        norms = np.linalg.norm(Wo, axis=1)
        top_indices = np.argsort(norms)[::-1][:num_tokens]
        
        top_tokens = [idx_to_token[idx] for idx in top_indices]
        top_norms = norms[top_indices]
        
        plt.figure(figsize=(10, 5))
        plt.bar(top_tokens, top_norms)
        plt.xlabel("Tokens")
        plt.ylabel("Weight Norm (Influence)")
        plt.title(f"Top {num_tokens} Influential Tokens in Output Layer (Wo)")
        plt.xticks(rotation=45, ha='right')
        plt.tight_layout()
        plt.show()

# Plot Attention Map for a Prompt with Generated Tokens
def plot_prompt_attention_map(model, prompt, generated_ids, vocab, idx_to_token, sequence_length, layer_idx=0, head_idx=0):
    model.eval()
    # Convert generated_ids to tokens
    full_sequence_ids = generated_ids[0].cpu().numpy().tolist()  # Full sequence including prompt and generated
    tokens = [idx_to_token[idx] for idx in full_sequence_ids]
    
    # Ensure tokens match sequence_length
    if len(tokens) > sequence_length:
        tokens = tokens[-sequence_length:]
    elif len(tokens) < sequence_length:
        tokens = ['<pad>'] * (sequence_length - len(tokens)) + tokens
    
    with torch.no_grad():
        _, attn_outputs = model(generated_ids, return_attn=True)
        attn_weights = attn_outputs[layer_idx][0, head_idx].cpu().numpy()  # (seq_len, seq_len)
    
    plt.figure(figsize=(10, 8))
    plt.imshow(attn_weights, cmap='viridis')
    plt.colorbar(label='Attention Weight')
    plt.xticks(range(sequence_length), tokens, rotation=45, ha='right')
    plt.yticks(range(sequence_length), tokens)
    plt.xlabel("Key Tokens")
    plt.ylabel("Query Tokens")
    plt.title(f"Attention Map (Layer {layer_idx}, Head {head_idx}) for Prompt: '{prompt}'")
    plt.tight_layout()
    plt.show()

# Main Execution
if __name__ == "__main__":
    # Initialize model
    model = GPTLikeModel(
        vocab_size=vocab_size,
        embed_dim=embed_dim,
        num_heads=num_heads,
        hidden_dim=hidden_dim,
        num_layers=num_layers,
        dropout=dropout,
        sequence_length=sequence_length
    ).to(device)
    
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.Adam(model.parameters(), lr=learning_rate)
    
    # Train the model
    losses = train_model(model, train_loader, criterion, optimizer, num_epochs)
    plot_loss(losses)
    
    # Plot vocabulary influence
    plot_vocab_influence(model, vocab, idx_to_token, num_tokens=20)
    
    # Test generation and plot attention map
    start_sequence = "Picasso is"
    generated_text, final_input_ids = generate_text(model, start_sequence, vocab, idx_to_token, k, sequence_length)
    print(f"Generated text: {start_sequence} {generated_text}")
    plot_prompt_attention_map(model, start_sequence, final_input_ids, vocab, idx_to_token, sequence_length, layer_idx=0, head_idx=0)
```
